```python

##Calcular los ingresos de cada tienda.
ingreso_Total1=tienda1['Precio'].sum()
ingreso_Total2=tienda2['Precio'].sum()
ingreso_Total3=tienda3['Precio'].sum()
ingreso_Total4=tienda4['Precio'].sum()
##Mostrar los resultados
print(f'Ingresos totales de la Primera tienda: {ingreso_Total1} ')
print(f'Ingresos totales de la Segunda tienda: {ingreso_Total2} ')
print(f'Ingresos totales de la Tercera tienda: {ingreso_Total3} ')
print(f'Ingresos totales de la Cuarta tienda: {ingreso_Total4} ')
```

```python
#Gráfica de los Ingresos Totales por Tienda

# Configuración de formato de números decimales
pd.options.display.float_format = '{:.2f}'.format

#Definir los datos que van en la tabla
tiendas= ['Tienda_1','Tienda_2','Tienda_3','Tienda_4']
ingresos_tiendas = [ingreso_Total1,ingreso_Total2,ingreso_Total3,ingreso_Total4]
#colores = ['#75ff33','#dbff33','#ffbd33','#ff5733']

# Paleta de colores en tonos de azul
colores = plt.cm.Blues(np.linspace(0.2, 1.0, len(tiendas)))


# Configuración de la figura
plt.figure(figsize=(9, 5))
plt.rcParams.update({'font.size': 12, 'font.family': 'serif'})

# Creación del gráfico de barras con bordes
barras = plt.bar(tiendas, ingresos_tiendas, color=colores, edgecolor='black', alpha=0.85)

# Etiquetas de ejes
plt.ylabel('Ingresos ($)', fontsize=12, fontweight='bold')
plt.xlabel('Tiendas', fontsize=12, fontweight='bold')
plt.title('Ingresos por Tienda', fontsize=14, fontweight='bold', color='#003f5c')

# Ajuste del fondo y cuadrícula
plt.gca().set_facecolor('#e6e6e6')  # Fondo más claro para mejor contraste
plt.grid(axis='y', linestyle='--', alpha=0.8, color='gray')  # Cuadrícula más visible

# Agregar etiquetas de valores sobre las barras
for barra in barras:
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval + (yval * 0.05),
             f"${round(yval,2):,.2f}", ha='center', va='bottom', fontsize=10,
             fontweight='bold', color='black')

# Ajuste del diseño y visualización
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Configuración del formato numérico
pd.options.display.float_format = '{:.2f}'.format

# Datos y cálculo de calificación promedio
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
dfs = [tienda1, tienda2, tienda3, tienda4]
calificaciones_promedio = [df['Calificación'].mean() for df in dfs]

# Ajustar el tamaño de los puntos según la calificación promedio
tamaño_puntos = [cal * 300 for cal in calificaciones_promedio]  # Mayor escala

# Configuración de la figura
plt.figure(figsize=(9, 5))

# Gráfico de dispersión con puntos escalados
colores = plt.cm.Blues(np.linspace(0.2, 1.0, len(tiendas)))
plt.scatter(tiendas, calificaciones_promedio, color=colores, edgecolors='black', s=tamaño_puntos, alpha=0.85, linewidths=2)

# Personalización del gráfico
plt.title('Calificación Promedio por Tienda', fontsize=14, fontweight='bold', color='#003f5c')
plt.xlabel('Tiendas', fontsize=12, fontweight='bold')
plt.ylabel('Calificación Promedio', fontsize=12, fontweight='bold')
plt.xticks(rotation=0, fontsize=10)
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.8, color='gray')

# Determinar el valor mínimo y máximo de calificaciones
min_calificacion = min(calificaciones_promedio)
max_calificacion = max(calificaciones_promedio)

# Agregar etiquetas con colores adecuados
for i, calificacion_valor in enumerate(calificaciones_promedio):
    if calificacion_valor == min_calificacion:
        color_etiqueta = '#FF0000'  # Rojo para el menor
        color_texto = 'white'
    elif calificacion_valor == max_calificacion:
        color_etiqueta = '#FFD700'  # Amarillo para el mayor
        color_texto = 'black'
    else:
        color_etiqueta = '#FFA500'  # Naranja para valores intermedios
        color_texto = 'black'

    plt.text(i, calificacion_valor + 0.3, f"{round(calificacion_valor,2):.2f}", ha='center', va='bottom',
             fontsize=12, fontweight='bold', color=color_texto,
             bbox=dict(facecolor=color_etiqueta, alpha=0.7, boxstyle='round,pad=0.3'))

# Ajuste del diseño
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display # Importar display

# Diccionario con nombres de tiendas y sus URLs
urls_tiendas = {
    'Tienda 1': url1,
    'Tienda 2': url2,
    'Tienda 3': url3,
    'Tienda 4': url4
}

# Configuración del formato numérico
pd.options.display.float_format = '{:.2f}'.format

# Iterar sobre cada tienda y generar el top 5 y bottom 5 de productos vendidos
for nombre_tienda, url in urls_tiendas.items():
    # Leer los datos
    df = pd.read_csv(url)

    # Calcular los productos más vendidos y menos vendidos
    # Usar reset_index() y luego renombrar las columnas
    top_5 = df['Producto'].value_counts().head(5).reset_index()
    top_5.columns = ['Producto', 'Cantidad Vendida'] # Renombrar columnas después de reset_index
    top_5 = top_5.sort_values(by='Cantidad Vendida', ascending=False) # Ordenar por el nuevo nombre de columna

    bottom_5 = df['Producto'].value_counts().tail(5).reset_index()
    bottom_5.columns = ['Producto', 'Cantidad Vendida'] # Renombrar columnas después de reset_index
    bottom_5 = bottom_5.sort_values(by='Cantidad Vendida', ascending=False) # Ordenar por el nuevo nombre de columna


    # Mostrar tablas con formato azul y orden correcto
    azul_fondo = '#003f5c'
    display(top_5.style.set_table_styles([
        {'selector': 'th', 'props': f'font-weight: bold; background-color: {azul_fondo}; color: white; text-align: center;'},
        {'selector': 'td', 'props': 'text-align: center; font-size: 12px;'},
    ]).set_properties(**{'border': '1px solid black', 'padding': '5px'}))

    display(bottom_5.style.set_table_styles([
        {'selector': 'th', 'props': f'font-weight: bold; background-color: {azul_fondo}; color: white; text-align: center;'},
        {'selector': 'td', 'props': 'text-align: center; font-size: 12px;'},
    ]).set_properties(**{'border': '1px solid black', 'padding': '5px'}))

    # Configuración del gráfico de barras
    plt.figure(figsize=(12, 6))

    # Colores diferenciados
    colores_top = plt.cm.Blues(np.linspace(0.3, 0.9, len(top_5)))
    colores_bottom = plt.cm.Blues(np.linspace(0.3, 0.9, len(bottom_5)))

    # Subgráfico de productos más vendidos
    plt.subplot(1, 2, 1)
    barras_top = plt.barh(top_5['Producto'], top_5['Cantidad Vendida'], color=colores_top, edgecolor='black', alpha=0.85)
    plt.xlabel('Cantidad Vendida')
    plt.title(f'Top 5 Más Vendidos - {nombre_tienda}', fontsize=14, fontweight='bold', color=azul_fondo)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.gca().invert_yaxis()  # Asegura el orden descendente en la gráfica
    for barra in barras_top:
        plt.text(barra.get_width() + 3, barra.get_y() + barra.get_height()/2, f"{barra.get_width():,.0f}",
                 ha='center', va='center', fontsize=10, fontweight='bold', color='black')

    # Subgráfico de productos menos vendidos
    plt.subplot(1, 2, 2)
    barras_bottom = plt.barh(bottom_5['Producto'], bottom_5['Cantidad Vendida'], color=colores_bottom, edgecolor='black', alpha=0.85)
    plt.xlabel('Cantidad Vendida')
    plt.title(f'Bottom 5 Menos Vendidos - {nombre_tienda}', fontsize=14, fontweight='bold', color=azul_fondo)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.gca().invert_yaxis()  # Asegura el orden descendente en la gráfica
    for barra in barras_bottom:
        plt.text(barra.get_width() + 3, barra.get_y() + barra.get_height()/2, f"{barra.get_width():,.0f}",
                 ha='center', va='center', fontsize=10, fontweight='bold', color='black')

    # Ajuste del diseño
    plt.tight_layout()
    plt.show()
```

# 5. Envío promedio por tienda

```python


# Configuración del formato numérico
pd.options.display.float_format = '{:.2f}'.format

# Diccionario con nombres de tiendas y sus DataFrames
tiendas_df = {
    'Tienda 1': tienda1,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}

# Calcular el costo de envío promedio por tienda
envios_promedio = {nombre: df['Costo de envío'].mean() for nombre, df in tiendas_df.items()}

# Convertir los resultados en un DataFrame
envios_df = pd.DataFrame(list(envios_promedio.items()), columns=['Tienda', 'Costo de Envío Promedio']).sort_values(by='Costo de Envío Promedio', ascending=False)

# Mostrar tabla con formato mejorado
azul_fondo = '#003f5c'
display(envios_df.style.set_table_styles([
    {'selector': 'th', 'props': f'font-weight: bold; background-color: {azul_fondo}; color: white; text-align: center;'},
    {'selector': 'td', 'props': 'text-align: center; font-size: 12px;'},
]).set_properties(**{'border': '1px solid black', 'padding': '5px'}))

# Crear gráfico de pastel con una gama de azules
plt.figure(figsize=(8, 8))
colores = plt.cm.Blues(np.linspace(0.3, 0.9, len(envios_df)))

# Gráfico de pastel con etiquetas mejoradas
plt.pie(envios_df['Costo de Envío Promedio'], labels=envios_df['Tienda'], colors=colores,
        autopct=lambda p: f'{p:.1f}%\n(${p * envios_df["Costo de Envío Promedio"].sum() / 100:.2f})',
        startangle=140, wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})

# Personalización del gráfico
plt.title('Distribución del Costo de Envío Promedio por Tienda', fontsize=14, fontweight='bold', color=azul_fondo)
plt.axis('equal')  # Mantiene la proporción circular
plt.show()
```
