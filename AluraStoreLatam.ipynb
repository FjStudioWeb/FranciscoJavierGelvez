```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Configuraci√≥n del formato num√©rico
pd.options.display.float_format = '{:.2f}'.format

# Datos y c√°lculo de calificaci√≥n promedio
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
dfs = [tienda1, tienda2, tienda3, tienda4]
calificaciones_promedio = [df['Calificaci√≥n'].mean() for df in dfs]

# Ajustar el tama√±o de los puntos seg√∫n la calificaci√≥n promedio
tama√±o_puntos = [cal * 300 for cal in calificaciones_promedio]  # Mayor escala

# Configuraci√≥n de la figura
plt.figure(figsize=(9, 5))

# Gr√°fico de dispersi√≥n con puntos escalados
colores = plt.cm.Blues(np.linspace(0.2, 1.0, len(tiendas)))
plt.scatter(tiendas, calificaciones_promedio, color=colores, edgecolors='black', s=tama√±o_puntos, alpha=0.85, linewidths=2)

# Personalizaci√≥n del gr√°fico
plt.title('Calificaci√≥n Promedio por Tienda', fontsize=14, fontweight='bold', color='#003f5c')
plt.xlabel('Tiendas', fontsize=12, fontweight='bold')
plt.ylabel('Calificaci√≥n Promedio', fontsize=12, fontweight='bold')
plt.xticks(rotation=0, fontsize=10)
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.8, color='gray')

# Determinar el valor m√≠nimo y m√°ximo de calificaciones
min_calificacion = min(calificaciones_promedio)
max_calificacion = max(calificaciones_promedio)

# Agregar etiquetas con colores adecuados
for i, calificacion_valor in enumerate(calificaciones_promedio):
    if calificacion_valor == min_calificacion:
        color_etiqueta = '#FF0000'  # Rojo para el menor
        color_texto = 'white'
    elif calificacion_valor == max_calificacion:
        color_etiqueta = '#FFD700'  # Amarillo para el mayor
        color_texto = 'black'
    else:
        color_etiqueta = '#FFA500'  # Naranja para valores intermedios
        color_texto = 'black'

    plt.text(i, calificacion_valor + 0.3, f"{round(calificacion_valor,2):.2f}", ha='center', va='bottom',
             fontsize=12, fontweight='bold', color=color_texto,
             bbox=dict(facecolor=color_etiqueta, alpha=0.7, boxstyle='round,pad=0.3'))

# Ajuste del dise√±o
plt.tight_layout()
plt.show()
```

# 4. Productos m√°s y menos vendidos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display # Importar display

# Diccionario con nombres de tiendas y sus URLs
urls_tiendas = {
    'Tienda 1': url1,
    'Tienda 2': url2,
    'Tienda 3': url3,
    'Tienda 4': url4
}

# Configuraci√≥n del formato num√©rico
pd.options.display.float_format = '{:.2f}'.format

# Iterar sobre cada tienda y generar el top 5 y bottom 5 de productos vendidos
for nombre_tienda, url in urls_tiendas.items():
    # Leer los datos
    df = pd.read_csv(url)

    # Calcular los productos m√°s vendidos y menos vendidos
    # Usar reset_index() y luego renombrar las columnas
    top_5 = df['Producto'].value_counts().head(5).reset_index()
    top_5.columns = ['Producto', 'Cantidad Vendida'] # Renombrar columnas despu√©s de reset_index
    top_5 = top_5.sort_values(by='Cantidad Vendida', ascending=False) # Ordenar por el nuevo nombre de columna

    bottom_5 = df['Producto'].value_counts().tail(5).reset_index()
    bottom_5.columns = ['Producto', 'Cantidad Vendida'] # Renombrar columnas despu√©s de reset_index
    bottom_5 = bottom_5.sort_values(by='Cantidad Vendida', ascending=False) # Ordenar por el nuevo nombre de columna


    # Mostrar tablas con formato azul y orden correcto
    azul_fondo = '#003f5c'
    display(top_5.style.set_table_styles([
        {'selector': 'th', 'props': f'font-weight: bold; background-color: {azul_fondo}; color: white; text-align: center;'},
        {'selector': 'td', 'props': 'text-align: center; font-size: 12px;'},
    ]).set_properties(**{'border': '1px solid black', 'padding': '5px'}))

    display(bottom_5.style.set_table_styles([
        {'selector': 'th', 'props': f'font-weight: bold; background-color: {azul_fondo}; color: white; text-align: center;'},
        {'selector': 'td', 'props': 'text-align: center; font-size: 12px;'},
    ]).set_properties(**{'border': '1px solid black', 'padding': '5px'}))

    # Configuraci√≥n del gr√°fico de barras
    plt.figure(figsize=(12, 6))

    # Colores diferenciados
    colores_top = plt.cm.Blues(np.linspace(0.3, 0.9, len(top_5)))
    colores_bottom = plt.cm.Blues(np.linspace(0.3, 0.9, len(bottom_5)))

    # Subgr√°fico de productos m√°s vendidos
    plt.subplot(1, 2, 1)
    barras_top = plt.barh(top_5['Producto'], top_5['Cantidad Vendida'], color=colores_top, edgecolor='black', alpha=0.85)
    plt.xlabel('Cantidad Vendida')
    plt.title(f'Top 5 M√°s Vendidos - {nombre_tienda}', fontsize=14, fontweight='bold', color=azul_fondo)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.gca().invert_yaxis()  # Asegura el orden descendente en la gr√°fica
    for barra in barras_top:
        plt.text(barra.get_width() + 3, barra.get_y() + barra.get_height()/2, f"{barra.get_width():,.0f}",
                 ha='center', va='center', fontsize=10, fontweight='bold', color='black')

    # Subgr√°fico de productos menos vendidos
    plt.subplot(1, 2, 2)
    barras_bottom = plt.barh(bottom_5['Producto'], bottom_5['Cantidad Vendida'], color=colores_bottom, edgecolor='black', alpha=0.85)
    plt.xlabel('Cantidad Vendida')
    plt.title(f'Bottom 5 Menos Vendidos - {nombre_tienda}', fontsize=14, fontweight='bold', color=azul_fondo)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.gca().invert_yaxis()  # Asegura el orden descendente en la gr√°fica
    for barra in barras_bottom:
        plt.text(barra.get_width() + 3, barra.get_y() + barra.get_height()/2, f"{barra.get_width():,.0f}",
                 ha='center', va='center', fontsize=10, fontweight='bold', color='black')

    # Ajuste del dise√±o
    plt.tight_layout()
    plt.show()
```

# 5. Env√≠o promedio por tienda

```python


# Configuraci√≥n del formato num√©rico
pd.options.display.float_format = '{:.2f}'.format

# Diccionario con nombres de tiendas y sus DataFrames
tiendas_df = {
    'Tienda 1': tienda1,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}

# Calcular el costo de env√≠o promedio por tienda
envios_promedio = {nombre: df['Costo de env√≠o'].mean() for nombre, df in tiendas_df.items()}

# Convertir los resultados en un DataFrame
envios_df = pd.DataFrame(list(envios_promedio.items()), columns=['Tienda', 'Costo de Env√≠o Promedio']).sort_values(by='Costo de Env√≠o Promedio', ascending=False)

# Mostrar tabla con formato mejorado
azul_fondo = '#003f5c'
display(envios_df.style.set_table_styles([
    {'selector': 'th', 'props': f'font-weight: bold; background-color: {azul_fondo}; color: white; text-align: center;'},
    {'selector': 'td', 'props': 'text-align: center; font-size: 12px;'},
]).set_properties(**{'border': '1px solid black', 'padding': '5px'}))

# Crear gr√°fico de pastel con una gama de azules
plt.figure(figsize=(8, 8))
colores = plt.cm.Blues(np.linspace(0.3, 0.9, len(envios_df)))

# Gr√°fico de pastel con etiquetas mejoradas
plt.pie(envios_df['Costo de Env√≠o Promedio'], labels=envios_df['Tienda'], colors=colores,
        autopct=lambda p: f'{p:.1f}%\n(${p * envios_df["Costo de Env√≠o Promedio"].sum() / 100:.2f})',
        startangle=140, wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})

# Personalizaci√≥n del gr√°fico
plt.title('Distribuci√≥n del Costo de Env√≠o Promedio por Tienda', fontsize=14, fontweight='bold', color=azul_fondo)
plt.axis('equal')  # Mantiene la proporci√≥n circular
plt.show()
```
